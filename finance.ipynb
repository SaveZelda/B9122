```python
# Generate increments using Euler maruyama approximation
def generateIncrements(t,n,mu,sigma,numPaths): 
    dt=t/n
    randomSamples=np.random.normal(0,1,(numPaths,n)) 
    X=mu*dt+sigma*math.sqrt(dt)*randomSamples 
    return(X)

def generateBM(X):
# Generate Brownian Motion
    numPaths=len(X)
    n=len(X[0])
    BM=np.zeros((numPaths,n+1))  
    for t in range(1,n+1):
        BM[:,t]=BM[:,t-1]+X[:,t-1] 
    return BM

def generate_exp_BM(X):
# Generate Brownian Motion
    numPaths=len(X)
    n=len(X[0])
    BM=np.zeros((numPaths,n+1)) 
    exp_BM=np.zeros((numPaths,n+1)) 
    for t in range(1,n+1):
        BM[:,t]=BM[:,t-1]+X[:,t-1]
        exp_BM[:,t] = np.exp(BM[:,t])
    return exp_BM
```

```python
increment = generateIncrements(1,252,0,1,100)
BM_1 = generateBM(increment)
for i in range (len(BM_1)):
    plt.plot(BM_1[i])
```

```python
print('The mean and stadnard error for 100 steps BM are',BM_1.mean(),BM_1.std()/np.sqrt(100))
```

```python
increment = generateIncrements(1,252,0,1,1000)
BM_2 = generateBM(increment)
for i in range (len(BM_2)):
    plt.plot(BM_2[i])
```

```python
print('The mean and standard error for 1000 steps BM are',BM_2.mean(),BM_2.std()/np.sqrt(1000))
```

```python
increment = generateIncrements(1,252,0,1,10000)
BM_3 = generateBM(increment)
for i in range (len(BM_3)):
    plt.plot(BM_3[i])
```

```python
print('The mean and standard error for 10000 steps BM are',BM_3.mean(),BM_3.std()/np.sqrt(10000))
```

```python
print('As the size of simulation increases, mean is closer to 0 and standard error is shrinking to the true value')
```

# Question 3b
Based on Euler maruyama approximation:
$$ B(t+\Delta t) = B(t) + \mu \Delta t + \sigma \sqrt(\Delta t) N(0,1)$$ has a expected value equals to 0 and variance equals to t when there is no drift in mean and diffusion in variance
$$ exp(B(t+\Delta t)) = \exp(0+0*(1/252)+1*\sqrt(1/252)*0) = \exp(0) = 1$$
Therefore, as the sample size increases, we should have a mean closer and closer to 0

```python
increment = generateIncrements(1,252,0,1,100)
BM_4 = generate_exp_BM(increment)
for i in range (len(BM_4)):
    plt.plot(BM_4[i])
```

```python
print('The mean and stadnard error for 100 steps BM are',BM_4.mean(),BM_4.std()/np.sqrt(100))
```

```python
increment = generateIncrements(1,252,0,1,1000)
BM_5 = generate_exp_BM(increment)
for i in range (len(BM_5)):
    plt.plot(BM_5[i])
```

```python
print('The mean and stadnard error for 1000 steps BM are',BM_5.mean(),BM_5.std()/np.sqrt(1000))
```

```python
print('As the size of simulation increases, mean is closer to e^0 = 1 and standard error is shrinking to the true value')
```
